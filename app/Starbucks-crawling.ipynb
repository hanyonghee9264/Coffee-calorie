## 콜드 브루 커피 크롤링 완성

```python
# parser 필요한 모듈 (selenium, bs4)
from selenium import webdriver
from bs4 import BeautifulSoup

# url 이미지 다운
import urllib.request

from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome('/Users/han-yonghee/Downloads/chromedriver')
driver.implicitly_wait(3)

# 스타벅스 URL 접근
# 현재페이지 반환 및 lxml로 파싱
driver.get('http://www.istarbucks.co.kr/menu/drink_list.do?CATE_CD=product_cold_brew')
html = driver.page_source
soup = BeautifulSoup(html, 'lxml')

# 콜드브루커피 리스트
product_list = soup.select_one('ul.product_cold_brew').select('li.menuDataSet')

# 콜드 브루 커피 이미지
cold_brew_image = soup.select_one('ul.product_cold_brew').select('li.menuDataSet dl dt a img')

count = 1
for img in cold_brew_image:
    cold_brew_image_url = img.get('src')
    urllib.request.urlretrieve(cold_brew_image_url, f'{count}.cold_brew_coffee.jpg')
    count += 1

# 콜드 브루 커피 8종 URL
cold_brew_url = []
for cold_brew in product_list:
    cold_brew_list = cold_brew.select('a')
    for cold_brew_prod in cold_brew_list:
        cold_brew_prod_number= cold_brew_prod['prod']
        cold_brew_baseurl = 'http://www.istarbucks.co.kr/menu/drink_view.do?product_cd='
        cold_brew_url.append(cold_brew_baseurl + cold_brew_prod_number)
        
# 콜드 브루 커피 8종 Detail 페이지
for url in cold_brew_url:
    try:
        driver.implicitly_wait(3)
        driver.get(url)
        html = driver.page_source
        soup = BeautifulSoup(html, 'lxml')
    except:
        print('{url} 오류')
        break
    # 콜드 브루 커피명
    cold_brew_name = soup.select_one('div.product_view_detail > div.myAssignZone > h4').get_text(strip=True)
    print(cold_brew_name)
    # 콜드 브루 커피 소개
    cold_brew_info = soup.select_one('div.product_view_detail > div.myAssignZone > .t1').get_text(strip=True)
    print(cold_brew_info)
    # 콜드 브루 커피 영양 단위
    cold_brew_nutrition_unit = soup.select_one('div.product_info_head > div.product_select_wrap2 > .selectTxt2').get_text(strip=True)
    print(cold_brew_nutrition_unit)
    # 콜드 브루 커피 칼로리
    cold_brew_calorie = soup.select_one('div.product_view_info > div.product_info_content > ul > li.kcal > dl > dd').get_text(strip=True)
    print(cold_brew_calorie)
    # 콜드 브루 커피 포화지방
    cold_brew_fat = soup.select_one('div.product_view_info > div.product_info_content > ul > li.sat_FAT > dl > dd').get_text(strip=True)
    print(cold_brew_fat)
    # 콜드 브루 커피 단백질
    cold_brew_protein = soup.select_one('div.product_view_info > div.product_info_content > ul > li.protein > dl > dd').get_text(strip=True)
    print(cold_brew_protein)
    # 콜드 브루 커피 나트륨
    cold_brew_sodium = soup.select_one('div.product_view_info > div.product_info_content > ul:nth-of-type(2) > li.sodium > dl > dd').get_text(strip=True)
    print(cold_brew_sodium)
    # 콜드 브루 커피 당류
    cold_brew_sugars = soup.select_one('div.product_view_info > div.product_info_content > ul:nth-of-type(2) > li.sugars > dl > dd').get_text(strip=True)
    print(cold_brew_sugars)
    # 콜드 브루 커피 카페인
    cold_brew_caffeine = soup.select_one('div.product_view_info > div.product_info_content > ul:nth-of-type(2) > li.caffeine > dl > dd').get_text(strip=True)
    print(cold_brew_caffeine)
    
driver.get('http://www.istarbucks.co.kr/menu/drink_list.do?CATE_CD=product_cold_brew')

driver.find_element_by_xpath('//*[@id="mCSB_1_container"]/li[2]/label').click()
driver.implicitly_wait(3)
driver.find_element_by_xpath('//*[@id="mCSB_1_container"]/li[3]/label').click()


```
